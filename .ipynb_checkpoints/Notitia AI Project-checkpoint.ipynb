```python
# Reading the EXCEL ffile using the col_names list in the names parameter:
data_df = pd.read_excel("notitia-ai-project.xlsx",index_col=0)
```

```python
# Fill NaN with Zeros
data_df = data_df.fillna(0)
```

```python
# Checking Correlation of Data
# # Pairplot using s 
corr = data_df.corr();
print(corr)
#sns.heatmap(corr, annot = True);

fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

plt.title('Heat Map of Correlation \n')

sns.heatmap(corr,annot=True,cmap='YlGnBu',
vmin=1,vmax=1,linewidths=.5, center=0);
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

data_df.groupby(data_df['Country'])['Adolescent_fertility_rate'].mean().plot.bar()
mean_birth = data_df["Adolescent_fertility_rate"].mean(0)
data_df["Adolescent_fertility_rate"].fillna(mean_birth)

# Creates labels
ax.set_xlabel('Country')
ax.set_ylabel('Adolescent fertility rate(Birth per 1000 women between age 15-19)')
# Sets the plot title
ax.set_title("Mean Adolescent_fertility_rate per Country")
ax.legend(bbox_to_anchor=(1, 1), loc=2, title='Indicator')

fig.tight_layout()
```
